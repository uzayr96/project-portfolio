## Converting Open Date to Date Format and Extracting Month, Year, Day and Day of the week

```python
## Remove Outlier
train = train[train['revenue'] < 15000000]
y = train['revenue']
n_train = train.shape[0]

## Drop revenue
train.drop('revenue', axis=1, inplace=True)

## Change MB to DT
test['Type'] = test['Type'].replace('MB', 'DT')

## Get Date Attributes
data = pd.concat([train, test], axis=0)
data['Date'] = pd.to_datetime(data['Open Date'], format='%m/%d/%Y')
data['Month'] = data['Date'].dt.month.astype(str)
data['Day'] = data['Date'].dt.day
data['Year'] = data['Date'].dt.year.astype(str)
data['Weekday'] = data['Date'].dt.day_name()
print("There are {} columns and {} rows in the dataset after getting date attributes".format(data.shape[1],
                                                                                               data.shape[0]))
```

```python
cols_to_remove_train = ['Open Date', 'Id', 'Date']

data_filtered = data.drop(cols_to_remove_train, axis=1)
print(data_filtered.shape)
```

```python
dummies = pd.get_dummies(data_filtered,drop_first=True)
print(dummies.shape)
dummies.head()
```

After getting dummies, I will remove special characters from the column names as xgboost algorithm will show error if the special characters are in the column names.

```python
dummies = dummies.rename(columns = lambda x:re.sub('[^A-Za-z0-9_]+', '', x))
X_train = dummies.iloc[:n_train]
X_test = dummies.iloc[n_train:]
print(X_train.shape)
print(X_test.shape)
(print(y.shape))
```

## Importing ML packages

```python
from sklearn.ensemble import RandomForestRegressor
from sklearn.model_selection import cross_val_score, RepeatedKFold
import xgboost as xgb
```

## Custom Functions
As the training dataset is small, I will use repeated KFold cross validation to find out how our model performs before predicting on our final test dataset. For that I will create a custom function that takes our model, train dataset and number of folds. I will also include a default parameter of columns to remove incase I use it in the training.

```python
## Score function
def score(model, train, y, nfolds, repeats,  columns_to_remove=None):
    kf = RepeatedKFold(n_splits=nfolds, random_state=42,n_repeats=repeats).get_n_splits(train.values)
    rmse = np.sqrt(-cross_val_score(model, train.values, y, scoring='neg_mean_squared_error',cv = kf))
    if columns_to_remove:
        train = train.drop(columns_to_remove, axis = 1)
        kf = RepeatedKFold(n_splits=nfolds, random_state=42,n_repeats=repeats).get_n_splits(train.values)
        rmse = np.sqrt(-cross_val_score(model, train.values, y, scoring='neg_mean_squared_error',cv = kf)) 
    return(rmse)

## trainmodel function
def trainModel(model, X, y, test,columns_to_remove=None):
    model.fit(X,y)
    pred = model.predict(test)
    if columns_to_remove:
        X = X.drop(columns_to_remove, axis = 1)
        model.fit(X, y)
        test = test.drop(columns_to_remove, axis = 1)
        pred = model.predict(test)
    
    return(pred)
```

```python
model_rf = RandomForestRegressor(n_estimators=100,
                                 max_depth=1,
                                 verbose=0,
                                 bootstrap=False,
                                 random_state=42,
                                 n_jobs=3,
                                 max_features=20)

print("[RANDOM FOREST] RMSE: ", score(model_rf,
                                      train = X_train,
                                      y=y,
                                      nfolds=10,
                                      repeats=5,
                                      columns_to_remove=None).mean())
```

The above parameters were randomly selected. The rmse of random forest from the cross validation score is around **160,000** and on the **public leaderboard(1% of the test data)** score we get around **183,000** and on the **private leaderboard (99% of the test data)** score we get around **182,000**.

```python
params = {
    'eta':0.1,
    'max_depth':1,
    'colsample_bytree':0.67,
    'min_child_weight':1,
    'objective':'reg:squarederror',
    'subsample':0.8,
    'reg_alpha':0.1,
    'reg_lambda':0.1,
    'n_estimators':25,
    'booster':'gbtree',
    'gamma':0
}



model_xgb = xgb.XGBRegressor(**params)



print("[XGboost] RMSE: ", score(model_xgb,
                                      train = X_train,
                                      y=y,
                                      nfolds=10,
                                      repeats=5,
                                      columns_to_remove=None).mean())
```

**Xgboost** on the public leaderboard performed way better than the randomforest but the private leaderboard result was quite similar.

```python
prediction_rf = trainModel(model_rf,X=X_train,y=y,test = X_test,columns_to_remove=None)
prediction_xgb = trainModel(model_xgb,X=X_train,y=y,test = X_test,columns_to_remove=None)
```

```python
submission = pd.DataFrame()
submission['Id'] = test['Id']
submission['Prediction'] = (prediction_xgb)
submission.head()
```
